```python
import urllib.request
import re
import nltk
from nltk import sent_tokenize
```

```python
page = urllib.request.urlopen('https://raw.githubusercontent.com/dashapetrova/nlp/master/text.txt')
text = page.read().decode('UTF-8')
```

Берем небольшой отрывок в 34 предложений

```python
part = text[:2359]
```

```python
part = re.sub("\n",'', part) #небольшая чистка
```

```python
part
```

```python
f = open('gold_example.txt', 'r', encoding = 'utf-8')
```

```python
gold = f.readlines()
```

```python
gold = [re.sub("\n",'', i) for i in gold] #небольшая чистка
```

```python
gold
```

Разделение регулярным выражением

```python
regex = r'(?<=[.!?]) (?=[A-ЯЁA-Z])'
```

```python
my_split_1 = re.split(regex, part)
```

```python
my_split_1
```

```python
print('Gold length - ', len(gold))
print('My_split_1 length- ', len(my_split_1))
```

```python
len(set(gold) & set(my_split_1)) / len(set(gold) | set(my_split_1))
```

На данном открывке разбиение регулярным выражением получилось корректным.

Посмотрим на sent_tokenize

```python
my_split_2 = sent_tokenize(part)
```

```python
my_split_2
```

```python
print('Gold length - ', len(gold))
print('My_split_2 length- ', len(my_split_2))
```

```python
len(set(gold) & set(my_split_2)) / len(set(gold) | set(my_split_2))
```

Можно заметить, что происходит ошибка в том месте, где есть сокращение имени "Д.Б."

Tokenizer

```python
from pprint import pprint
from nltk.tokenize.punkt import PunktSentenceTokenizer, PunktTrainer
```

```python
trainer = PunktTrainer()
trainer.INCLUDE_ALL_COLLOCS = True
trainer.train(text)
 
tokenizer = PunktSentenceTokenizer(trainer.get_params())
```

```python
print(tokenizer._params.abbrev_types)
```

tokenizer распознал сокращение "д.б"

```python
tok = tokenizer.tokenize(part)
```

```python
tok
```

```python
print('Gold length - ', len(gold))
print('Tok length- ', len(tok))
```

```python
len(set(gold) & set(tok)) / len(set(gold) | set(tok))
```

Разбиение данного отрывка получилось корректным. Ошибок для исправления нет.
